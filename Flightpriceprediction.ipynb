```python
#can also use lambda function
df['Date']=df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
df['Month']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
df['Year']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df.isnull().sum()
```

```python
df['Arrival_Hour']=df['Arrival_Time'].apply(lambda x:x.split(':')[0]).astype(int)
df['Arrival_Min']=df['Arrival_Time'].apply(lambda x:x.split(':')[1]).astype(int)
```
